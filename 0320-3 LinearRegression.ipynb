```python
import pandas as pd

df_time = pd.read_csv('C:/부동산/data/df_time.csv')
```

```python
df_time
```

```python
df_time.set_index('년월', inplace=True)
```

```python
df_time
```

```python
import statsmodels.formula.api as smf

model = smf.ols(formula = '''면적당가격 ~ 미국달러 + 유로
+ 소비자물가지수 + KOSDAQ_종가
+ 무담보콜금리 + KORIBOR + 국고채_1년''', data=df_time).fit()
model.summary()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
X = df_time[['미국달러', '유로', '소비자물가지수',
       'KOSDAQ_종가', '무담보콜금리', 'KORIBOR', '국고채_1년']]
y = df_time[['면적당가격']]
model.fit(X, y)
```

```python
미국달러 = 1000
유로 = 1200
소비자물가지수 = 80
KOSDAQ_종가 = 700
무담보콜금리 = 5
KORIBOR = 5
국고채_1년 = 5

timeA = [미국달러, 유로, 소비자물가지수, KOSDAQ_종가, 무담보콜금리, KORIBOR, 국고채_1년]
valueA = model.predict([timeA])
valueA
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
model = LinearRegression().fit(X_train, y_train)
```

```python
print("학습용 :", model.score(X_train, y_train))
print("검증용 :", model.score(X_test, y_test))
```
