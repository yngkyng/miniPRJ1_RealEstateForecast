```python
#데이터 불러오기
df = pd.read_excel("C:/Users/tjoeun/Desktop/datafile/2007(1).xlsx")
```

```python
# 앞의 수식어 설명 제거
df.drop(range(15), axis=0, inplace =True)
```

```python
# 첫번째 행을 column으로 지정
df2 = df.rename(columns = df.iloc[0])
```

```python
# 컬럼 지정후 남는 행 지우기
df2.drop([15], axis=0,inplace =True)
```

```python
# 인덱스 순서 초기화
df2 = df2.set_index('시군구').reset_index()
```

```python
df3=df2[['시군구','전용면적(㎡)','거래금액(만원)','층','건축년도']]
```

```python
a = df2['계약년월']
b = df2['계약일']
c = []
for i in range(len(b)):
    if int(b[i]) <10:
        ans = str(a[i])+str('0')+str(b[i])
        c.append(ans)
    else:
        ans = str(a[i])+str(b[i])
        c.append(ans)
```

```python
df3['계약년월일'] = c
```
